```python
plt.plot(x,y)
plt.title("This Title");
```

```python
plt.plot(x, y)
plt.title("This Title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show();
```

```python
plt.plot(x, y)
plt.title("this title")
plt.xlabel("x axis")
plt.ylabel("y axis")
# adding the limitation of x and y axis
plt.xlim(0,8)
plt.ylim(0,15)
plt.show();
```

```python
plt.plot(x, y);
plt.title("this title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(0,8)
plt.ylim(0,15)
plt.savefig("my_figure.png")
```

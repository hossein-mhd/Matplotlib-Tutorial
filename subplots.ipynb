```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = x*2
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)

#fig.tight_layout()
# .tight_layout() fixes the space between the figures problem

# OR

fig.subplots_adjust(wspace=0.9, hspace=0.5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][0].set_xlabel('x 0,0')

axes[0][1].plot(a, b)
axes[0][1].set_title('TITLE 0,1')

axes[1][0].plot(a, b)

axes[1][1].plot(a, b)

fig.suptitle('SUPER TITLE', fontsize=16)
# this the main title for all figures

fig.subplots_adjust(wspace=0.9, hspace=0.5)

#fig.savefig('2dimen_fig.png', bbox_inches='tight')
```

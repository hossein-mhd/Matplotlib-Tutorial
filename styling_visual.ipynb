```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color="red", lw = 5)
ax.plot(x,x+1, color='#e6c017', linewidth = 10)
# lw is equal to linewidth, there is no defferent
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color="red", lw = 6, ls = '--')
ax.plot(x,x+1, color='#e6c017', linewidth = 3, linestyle = '-.')
# ls is equal to linestyle, there is no defferent
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x+1, color='#e6c017', linewidth = 3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x, color='#e6c017', linewidth = 2, marker="o", ms = 10)
lines = ax.plot(x,x+1, color='black', linewidth = 1, marker="+", markersize=20)
# ms is equal to markersize, there is no defferent
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x, color='black', linewidth = 3, marker="o",ls="--", markersize=20,
               markerfacecolor="black", markeredgecolor="#e6c017", markeredgewidth=4)
```

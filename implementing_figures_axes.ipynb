```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)
```

```python
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# small axes
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(x, y)
```

```python
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('power of 4')

# small axes
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])

axes2.set_xlim(1,2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('zoomed in')

axes2.plot(x, y)
```

```python
fig = plt.figure(figsize=(10,3), dpi=200)
# dpi makes it more clear but also it takes more ram

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

plt.savefig('new_fig2.jpeg')
plt.savefig('C:\\Users\\Chalus System\\Desktop\\new_fig2.jpeg',bbox_inches='tight')
```
